```python
# Import Libs
import pandas as pd
```

Day 01

```python
# Extract Data

# PARQUET
raw_link = 'https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/raw/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet'

df_parquet = pd.read_parquet(raw_link)

```

```python
# CSV

anos = [20101, 20111, 20121, 20131, 20141, 20151, 20161, 20171, 20181, 20191, 20201, 20102, 20112, 20122, 20132, 20142, 20152, 20162, 20172, 20182, 20192]

# Lista para armazenar todos os DataFrames lidos
dfs = []

for year in anos:
    csv_link = f'https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/raw/main/Dia_1-Importando_dados/Datasets/dados_emprestimos/emprestimos-{year}.csv'
    df_csv = pd.read_csv(csv_link)
    dfs.append(df_csv)
```

```python
# Transformando e saneando a base

# Concatenando todos os DataFrames
df_csv_concat = pd.concat(dfs, ignore_index=True)

# Verificando e removendo dados duplicados
df_csv_concat.drop_duplicates(inplace=True)

print(df_csv_concat.head())

```

```python
# PARQUET

# Verificando e removendo dados duplicados
df_parquet.drop_duplicates(inplace=True)
```

```python
# Merge Parquet CSV


base_full = df_parquet.merge(df_csv_concat)
base_full
```
