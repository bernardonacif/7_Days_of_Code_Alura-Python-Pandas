```python
# Import Libs
import pandas as pd
```

Day 01

```python
# Extract Data

# PARQUET
raw_link = 'https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/raw/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet'

df_parquet = pd.read_parquet(raw_link)

```

```python
# CSV

anos = [20101, 20111, 20121, 20131, 20141, 20151, 20161, 20171, 20181, 20191, 20201, 20102, 20112, 20122, 20132, 20142, 20152, 20162, 20172, 20182, 20192]

# Lista para armazenar todos os DataFrames lidos
dfs = []

for year in anos:
    csv_link = f'https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/raw/main/Dia_1-Importando_dados/Datasets/dados_emprestimos/emprestimos-{year}.csv'
    df_csv = pd.read_csv(csv_link)
    dfs.append(df_csv)
```

```python
# Transformando e saneando a base

# Concatenando todos os DataFrames
df_csv_concat = pd.concat(dfs, ignore_index=True)

# Verificando e removendo dados duplicados
df_csv_concat.drop_duplicates(inplace=True)

print(df_csv_concat.head())

```

```python
# PARQUET

# Verificando e removendo dados duplicados
df_parquet.drop_duplicates(inplace=True)
```

```python
# Merge Parquet CSV


base_full = df_parquet.merge(df_csv_concat)
base_full
```

Day 02

```python
# Criando Index_cdu

cdu_list = []
for i in base_full['localizacao']:
  if(i < 100):
    cdu_list.append('Generalidades')
  elif(i < 200):
    cdu_list.append('Filosofia e psicologia')
  elif(i < 300):
    cdu_list.append('Religião')
  elif(i < 400):
    cdu_list.append('Ciências sociais')
  elif(i < 500):
    cdu_list.append('Classe vaga')
  elif(i < 600):
    cdu_list.append('Matemática e ciências naturais')
  elif(i < 700):
    cdu_list.append('Ciências aplicadas')
  elif(i < 800):
    cdu_list.append('Belas artes')
  elif(i < 900):
    cdu_list.append('Linguagem')
  else:
    cdu_list.append('Geografia. Biografia. História.')
```

```python
base_full['cdu'] = cdu_list
base_full.head()
```

```python
# Drop coluna 'registro_sistema'
base_full.drop(columns=['registro_sistema'],inplace=True)
```

```python
#  Ajuste tipo matricula
base_full['matricula_ou_siape'] = base_full['matricula_ou_siape'].astype('string')
```

```python
# Analise grafica dos emprestimos 

# Anual

base_full['data_emprestimo'] = pd.to_datetime(base_full['data_emprestimo'])
base_full['ano_emprestimo'] = base_full['data_emprestimo'].dt.year
contagem_por_ano = base_full['ano_emprestimo'].value_counts()
contagem_por_ano = contagem_por_ano.sort_index()

print(contagem_por_ano)
```

```python
# Gerando grafico
contagem_por_ano.plot(kind='line', marker='o', linestyle='-')

plt.title('Quantidade de empréstimos por ano')
plt.xlabel('Ano')
plt.ylabel('Quantidade de empréstimos')

plt.grid(True)
plt.show()
```

```python
# Analise mensal

base_full['data_emprestimo'] = pd.to_datetime(base_full['data_emprestimo'])
base_full['mes_emprestimo'] = base_full['data_emprestimo'].dt.month_name()
contagem_por_mes = base_full['mes_emprestimo'].value_counts()

# print(contagem_por_mes)

ordem_meses = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
contagem_por_mes = contagem_por_mes.reindex(ordem_meses)

# Imprimir o resultado
print(contagem_por_mes)
```

```python
contagem_por_mes.plot(kind='line', marker='o', linestyle='-')

plt.title('Quantidade de empréstimos por mes')
plt.xlabel('Ano')
plt.ylabel('Quantidade de empréstimos')

plt.grid(True)
plt.show()
```

```python
# por hora

base_full['data_emprestimo'] = pd.to_datetime(base_full['data_emprestimo'])
base_full['hora_emprestimo'] = base_full['data_emprestimo'].dt.hour
contagem_por_hora = base_full['hora_emprestimo'].value_counts().sort_index()

print(contagem_por_hora)
```

```python
# Grafico por hora

contagem_por_hora.plot(kind='line', marker='o', linestyle='-')

plt.title('Quantidade de empréstimos por hora')
plt.xlabel('Ano')
plt.ylabel('Quantidade de empréstimos')

plt.grid(True)
plt.show()
```

```python
# Analisando o DF base_full

base_full_columns = []

for i in base_full.columns:
    base_full_columns.append(i)

# print(base_full_columns)

# Verificando tipo de dados de cada coluna

for column_name in base_full_columns:
    column_type = base_full[column_name].dtype
    print(f"A coluna '{column_name}' é do tipo: {column_type}")

```

```python
# Distict para cada coluna do DF

for column_name in base_full_columns:
    unique_values = base_full[column_name].unique()
    print(f'{column_name}\n')
    print(unique_values) 
    print(f'\n\n')

```

```python
#  criar funcao para calcular qtd vs % das colunas chaves

def calc_df(coluna):
    df_calc = pd.DataFrame(base_full[coluna].value_counts()) 
    df_calc.columns = ['qtd']
    df_calc['%'] = round((df_calc.qtd / df_calc.qtd.sum())*100,1)
    return df_calc
```

```python
# Gerando DF calculado para cada coluna

for i in base_full_columns:
    calc_df(i)
    print(f'calc_df("{i}")')
```
